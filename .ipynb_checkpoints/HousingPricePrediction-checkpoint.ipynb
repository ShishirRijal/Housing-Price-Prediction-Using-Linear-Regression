```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import copy
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

<h1>1. Read the dataset</h1>

```python
# read the csv file
def get_dataset(file):
    return pd.read_csv(file)

def read_csv(file):
    dataset = pd.read_csv(file)
    price = dataset['price']/1000 # in 1000 USD
    area = dataset['area']
    bedrooms = dataset['bedrooms']
    bathrooms = dataset['bathrooms']
    stories = dataset['stories']
    mainroad = dataset['mainroad']
    guestroom = dataset['guestroom']
    basement = dataset['basement']
    hotwater = dataset['hotwaterheating']
    ac = dataset['airconditioning']
    parking = dataset['parking']
    pref_area = dataset['prefarea']
    furnishing_status = dataset['furnishingstatus']

    return price, area, bedrooms, bathrooms, stories, mainroad, guestroom, basement, hotwater, ac, parking, pref_area, furnishing_status
```

```python
# read data from csv file to variables
dataset = get_dataset("Housing.csv")
price, area, bedrooms, bathrooms, stories, mainroad, guestroom, basement, hotwater, ac, parking, pref_area, furnishing_status = read_csv("Housing.csv")
```

```python
# Let's take a look at the data 
df = copy.deepcopy(dataset)
dataset
```

<h1>Data Analaysis</h1>

```python
# create a box-plot
figure, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in df.items():
    sns.boxplot(y=col, data=df, ax=ax[index] )
    index+=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# create hist plot
figure, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in df.items():
    sns.histplot(data=value, ax=ax[index] )
    index+=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# Converting yes and no to 1 and 0
cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

for col in cols:
    for index, val in enumerate(df[col]):
        if val == 'yes':
            df.at[index, col] = 1
        else:
            df.at[index, col] = 0
df.drop(columns=['furnishingstatus'], inplace=True)
# Visualize updated data 
df
```

```python
# Converting yes and no to 1 and 0
cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

for col in cols:
    for index, val in enumerate(df[col]):
        if val == 'yes':
            df.at[index, col] = 1
        else:
            df.at[index, col] = 0
df.drop(columns=['furnishingstatus'], inplace=True)
# Visualize updated data 
df
```

<h2>Maximum Minimum</h2>

```python
# Feature scaling the price and area 
cols = ['price', 'area']

for col in cols:
    max = np.max(df[col])
    min = np.min(df[col])
    df[col] = (df[col]-min)/(max-min)

df
    
```

```python
# Splitting the data set to training and test set
x = df.drop(['price'], axis=1)
y = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```
